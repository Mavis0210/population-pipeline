```python
import pandas as pd
```

```python
from sqlalchemy import create_engine
```

```python
data_url ='https://raw.githubusercontent.com/datasets/population/main/data/population.csv'
df = pd.read_csv(data_url)
```

```python
df.head()
```

```python
df.dtypes
```

```python
engine = create_engine('postgresql+psycopg2://root:root@localhost:5432/population_db')
```

```python
df.to_sql(name='world_population', con=engine, if_exists='replace')
```
