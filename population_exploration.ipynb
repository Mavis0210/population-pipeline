```python
import os
import pandas as pd
from sqlalchemy import create_engine
```

```python
data_url = 'https://raw.githubusercontent.com/datasets/population/main/data/population.csv'
df = pd.read_csv(data_url)
print(f"Shape: {df.shape}")
df.head()
```

```python
print(df.dtypes)
print(f"\nMissing values:\n{df.isnull().sum()}")
print(f"\nDuplicates: {df.duplicated().sum()}")
```

```python
df.columns = df.columns.str.lower().str.replace(' ', '_')
df = df.dropna()
df = df.drop_duplicates()
print(f"Clean shape: {df.shape}")
df.head()
```

```python
engine = create_engine(
    f"postgresql+psycopg2://root:root@localhost:5433/population_db"
)

df.to_sql(
    name='world_population',
    con=engine,
    if_exists='replace',
    index=False
)

count = pd.read_sql("SELECT COUNT(*) FROM world_population", engine).iloc[0, 0]
print(f"Done! {count} rows loaded into world_population")
```
